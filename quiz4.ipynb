### Punto 1

```python
import matplotlib.pyplot as plt


nombres = []
precios = []

for i in range(5):
    nombre_i = input(f'Dígite el nombre de su snack #{i + 1}: ')
    print()
    precio_i = int(input(f'Dígite el precio de su snack #{i + 1}: '))
    print()
    nombres.append(nombre_i)
    precios.append(precio_i)


plt.bar(nombres, precios)
plt.title('Snacks y precios')
plt.xlabel('Snack')
plt.ylabel('Precio')
plt.savefig('snacks_y_precios.png')
```

### Punto 2

```python
import matplotlib.pyplot as plt


ciudades = []
poblaciones = []
mayor_poblacion = 0
ciudad_mayor_poblacion = 0
for i in range(5):
    ciudad_i = input(f'Dígite el nombre de su ciudad favorita #{i + 1}: ')
    print()
    poblacion_i = float(input('Dígite su población: '))
    print()
    ciudades.append(ciudad_i)
    poblaciones.append(poblacion_i)
    if poblacion_i > mayor_poblacion:
        mayor_poblacion = poblacion_i
        ciudad_mayor_poblacion = i


print(f'Ciudad con mayor población: {ciudades[ciudad_mayor_poblacion]}')

explode = [0] * 5
explode[ciudad_mayor_poblacion] = 0.1
fig1, ax1 = plt.subplots()
ax1.pie(poblaciones, explode=explode, labels=ciudades, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title('Ciudades y poblaciones')
plt.savefig('ciudades_y_poblaciones.png')
```
